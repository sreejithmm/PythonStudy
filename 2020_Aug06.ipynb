```python
# %load https://tinyurl.com/fract-ex
"""
Implement the Fraction class below passing all
the test cases in the doc-tests.

    >>> a = Fraction(1, 2)
    >>> b = Fraction("1/4")
    >>> print(a)
    1/2
    >>> print(b)
    1/4
    >>> print(a + b)
    3/4
    >>> print(a - b)
    1/4
    >>> print(a * b)
    1/8
    >>> print(a / b)
    2/1
    >>> float(a)
    0.5
    >>> float(b)
    0.25
    >>> c = Fraction(33, 99)
    >>> print(c)
    1/3
    
"""
import math

class Fraction:
    """
    This class implements fractional numbers.
    If instantiated with two arguments in the constructor
    expression, each argument must be an integer representing
    the numerator and the denominator. If the constructor
    expession is passed with single argument, then the argument
    must be a string of the form "n/d" when n represents numerator
    and d represents denominator.

    For example:  
        >>> a = Fraction(1, 2) # Create a fraction 1/2
        >>> b = Fraction("1/4") # Create a fraction 1/4
        >>> c = a + b   # Add the two fractions
        >>> print(c)    # Print the string representation of c
        3/4
        >>> a > b
        True
        >>> c == b
        False
        


    """
    def __init__(self,numerator,denominator=None):
        if(type(numerator) is str):
            self.num,self.denom = [int(e) for e in (numerator.split('/'))]
            
        else:
            self.num = numerator
            self.denom = denominator
        gcd = math.gcd(self.num,self.denom)
        self.num = int(self.num/gcd)
        self.denom = int(self.denom/gcd)

    
    def __str__(self):
            return(f"{self.num}/{self.denom}")
    def __add__(self,other):
        import math
        nume = self.num * other.denom + other.num * self.denom
        denomi = self.denom * other.denom
        return(Fraction(nume,denomi))
    def __sub__(self,other):
        import math
        nume = self.num * other.denom - other.num * self.denom
        denomi = self.denom * other.denom
        return(Fraction(nume,denomi))
    def __mul__(self,other):
        import math
        nume = self.num * other.num 
        denomi = self.denom * other.denom
        return(Fraction(nume,denomi))

    def __truediv__(self,other):
        import math
        nume = self.num * other.denom 
        denomi = self.denom * other.num
        return(Fraction(nume,denomi))
    def __float__(self):
        return(float(self.num / self.denom))
    def __gt__(self,other):
        val1 = float(self.num/self.denom)
        val2 = float(other.num/other.denom)
        return (val1 > val2)


if __name__ == '__main__':
    import doctest
    doctest.testmod()

    a = Fraction(1, 2)
    print(a)
    b = Fraction("1/4")
    print(b)
    print(a + b)
    print(a - b)
    print(a * b)
    print(a / b)
    print(float(a))
    print(float(b))
    c = Fraction(33, 99)
    print(c)
```
